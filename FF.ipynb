```python
url_data = 'https://raw.githubusercontent.com/vaastav/Fantasy-Premier-League/master/data/2022-23/cleaned_players.csv'
url_excel = '/Users/joethorpe/Documents/Fantasy Football/FF_2023.xlsx'
```

```python
data = pd.DataFrame(pd.read_csv(url_data))
data
```

```python
#data.to_excel(url_excel, sheet_name='Data Input')

book = load_workbook(url_excel)

writer = pd.ExcelWriter(url_excel, engine='openpyxl') 

writer.book = book

writer.sheets = dict((ws.title, ws) for ws in book.worksheets)

data.to_excel(writer, sheet_name="Data_Input")

writer.save()
```
